This shows that while the Original Score reflects overall well-being, the Composite Score highlights the impact of economic and health factors on perceived happiness.

```
#3.1 Function to detect outliers
def find_outliers(column):
    Q1 = df_south[column].quantile(0.25)
    Q3 = df_south[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df_south[(df_south[column] < lower) | (df_south[column] > upper)]

outliers_score = find_outliers('score')
outliers_gdp = find_outliers('Log GDP per capita')

print("Outliers Based on Score:\n")
print(outliers_score[['Country name', 'score']])

print("\nOutliers Based on GDP per Capita:\n")
print(outliers_gdp[['Country name', 'Log GDP per capita']])
```

```
plt.figure(figsize=(8,6))
plt.scatter(df_south["Log GDP per capita"], df_south["score"], label="Normal")

# Plot Score outliers in red
plt.scatter(outliers_score["Log GDP per capita"], outliers_score["score"],
            s=120, edgecolors="black", label="Score Outlier")

# Plot GDP outliers in green
plt.scatter(outliers_gdp["Log GDP per capita"], outliers_gdp["score"],
            s=120, edgecolors="black", label="GDP Outlier")

plt.xlabel("GDP per Capita")
plt.ylabel("Score")
plt.title("Outliers in South Asia (GDP vs Score)")
plt.legend()
plt.show()
```

In the scatter plot of the South Asian countries with GDP per Capita against Happiness Score, we can identify two outliers. The former, a score outlier (orange point), has much higher Happiness Score than other countries of the region, although it has an average GDP per capita. This would imply that its excessive score could be due to social support, health or governance. Its existence may exaggerate the average Happiness Score in the region, and increase the appearance of a happier region than most countries actually are. The second one is GDP outlier (green point) that has a very low GDP per capita, which is significantly lower than its neighbors and still has a medium level of Happiness Score. It means that despite the low economic output, there can be social or cultural reasons to maintain the decent level of happiness. This outlier may reduce the average GDP per capita and may cause inferences of the economy of the regions. On the whole, these outliers point out the extraordinary cases that are not typical and indicate that the use of only mean values might represent the region inaccurately, which is the reason why median or robust statistics might be a better attempt to create the image.
